```python
#import all libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as war
war.filterwarnings('ignore')
```

```python
#read ipl.csv file and print first 5 records
ds=pd.read_csv('AustraliaDataScienceJobs.csv')
ds.head()
```

```python
ds.info()
```

```python
ds.describe()
```

```python
# find null values 
ds.isnull().sum()/len(ds)*100
```

```python
#Numerical Columns
num_col=ds.select_dtypes(include=['int','float']).columns
num_col
```

```python
#Catogorical Columns
cat_col=ds.select_dtypes(include=['object']).columns
cat_col
```

```python
ds.shape
```

```python
ds.dropna(subset=['Job Title'],inplace=True)
```

```python
ds['Company Size']=ds['Company Size'].fillna(ds['Company Size'].mode()[0])
ds['Company Type']=ds['Company Type'].fillna(ds['Company Type'].mode()[0])
ds['Company Sector']=ds['Company Sector'].fillna(ds['Company Sector'].mode()[0])
ds['Company Industry']=ds['Company Industry'].fillna(ds['Company Industry'].mode()[0])
ds['Company Revenue']=ds['Company Revenue'].fillna(ds['Company Revenue'].mode()[0])
```

```python
ds.isnull().sum()/len(ds)*100
```

```python
ds['Company Rating']=ds['Company Rating'].fillna(ds['Company Rating'].median())
ds['Company Friend Reccomendation']=ds['Company Friend Reccomendation'].fillna(ds['Company Friend Reccomendation'].median())
ds['Company CEO Approval']=ds['Company CEO Approval'].fillna(ds['Company CEO Approval'].median())
ds['Companny Number of Rater']=ds['Companny Number of Rater'].fillna(ds['Companny Number of Rater'].median())
ds['Company Career Opportinities']=ds['Company Career Opportinities'].fillna(ds['Company Career Opportinities'].median())
ds['Compensation and Benefits']=ds['Compensation and Benefits'].fillna(ds['Compensation and Benefits'].median())
ds['Company Culture and Values']=ds['Company Culture and Values'].fillna(ds['Company Culture and Values'].median())
ds['Company Senior Management']=ds['Company Senior Management'].fillna(ds['Company Senior Management'].median())
ds['Company Work Life Balance']=ds['Company Work Life Balance'].fillna(ds['Company Work Life Balance'].median())
```

```python
ds['Company Founded']=ds['Company Founded'].fillna(ds['Company Founded'].mode()[0])
```

```python
ds.isnull().sum()/len(ds)*100
```

```python
ds['Company Founded']=ds['Company Founded'].astype("int")
```
